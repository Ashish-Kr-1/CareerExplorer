```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
def split(obj):
    if '+' in obj:
        return obj.split('+')
    elif '/' in obj:
        return obj.split('/')
    elif ',' in obj:
        return obj.split(',')
    else:
        return [obj]
```

```python
def strip(lst):
    l=[]
    for i in lst:
        if ',' in i:
            l.append(i.strip(','))
        else:
            l.append(i.strip())
    return l
```

```python
df["Qualification"] = df["Qualification"].apply(split)
```

```python
df["Qualification"] = df["Qualification"].apply(strip)
```

```python
df["Skills"] = df["Skills"].apply(split)
```

```python
df["Skills"] = df["Skills"].apply(strip)
```

```python
import json

# Input and output file paths
ldjson_file = "decsription.ldjson"
json_file = "description.json"

# Read the LDJSON file and convert to JSON
with open(ldjson_file, "r") as infile:
    data = [json.loads(line) for line in infile]

# Write the data to a JSON file
with open(json_file, "w") as outfile:
    json.dump(data, outfile, indent=4)

print(f"Converted LDJSON to JSON and saved as {json_file}")
```
