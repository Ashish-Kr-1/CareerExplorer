```python
def split(obj):
    if '+' in obj:
        obj = obj.replace('+',' ')
        return obj.split()
    elif '/' in obj:
        obj = obj.replace('/',' ')
        return obj.split()
    elif ',' in obj:
        obj = obj.replace(',',' ')
        return obj.split()
    else:
        return obj.split()
```

```python
def strip(lst):
    l=[]
    for i in lst:
        if ',' in i:
            l.append(i.strip(','))
        else:
            l.append(i.strip())
    return l
```

```python
df["Qualification"] = df["Qualification"].apply(split)
```

```python
df["Qualification"] = df["Qualification"].apply(strip)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['Tags'] = new_df['Tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=200, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['Tags']).toarray()
```
